```python
import yahoo_fin.stock_info as si
import pandas as pd
import re
```

```python
aapl_earnings_hist_df = pd.DataFrame.from_dict(si.get_earnings_history("aapl"))
meta_earnings_hist_df = pd.DataFrame.from_dict(si.get_earnings_history("meta"))
goog_earnings_hist_df = pd.DataFrame.from_dict(si.get_earnings_history("goog"))
nflx_earnings_hist_df = pd.DataFrame.from_dict(si.get_earnings_history("nflx"))
amzn_earnings_hist_df = pd.DataFrame.from_dict(si.get_earnings_history("amzn"))
msft_earnings_hist_df = pd.DataFrame.from_dict(si.get_earnings_history("msft"))
```

```python
years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
```

```python
aapl_earnings_10_yr_df = pd.Series(aapl_earnings_hist_df[aapl_earnings_hist_df['startdatetime'].str.contains('|'.join(years))].replace('(T(.*?)Z)','', regex=True, inplace=False)['startdatetime']).rename('aapl')
meta_earnings_10_yr_df = pd.Series(meta_earnings_hist_df[meta_earnings_hist_df['startdatetime'].str.contains('|'.join(years))].replace('(T(.*?)Z)','', regex=True, inplace=False)['startdatetime']).rename('meta')
goog_earnings_10_yr_df = pd.Series(goog_earnings_hist_df[goog_earnings_hist_df['startdatetime'].str.contains('|'.join(years))].replace('(T(.*?)Z)','', regex=True, inplace=False)['startdatetime']).rename('goog')
nflx_earnings_10_yr_df = pd.Series(nflx_earnings_hist_df[nflx_earnings_hist_df['startdatetime'].str.contains('|'.join(years))].replace('(T(.*?)Z)','', regex=True, inplace=False)['startdatetime']).rename('nflx')
amzn_earnings_10_yr_df = pd.Series(amzn_earnings_hist_df[amzn_earnings_hist_df['startdatetime'].str.contains('|'.join(years))].replace('(T(.*?)Z)','', regex=True, inplace=False)['startdatetime']).rename('amzn')
msft_earnings_10_yr_df = pd.Series(msft_earnings_hist_df[msft_earnings_hist_df['startdatetime'].str.contains('|'.join(years))].replace('(T(.*?)Z)','', regex=True, inplace=False)['startdatetime']).rename('msft')
```

```python
earnings_10yr_df = pd.concat(
    [
        aapl_earnings_10_yr_df, 
        meta_earnings_10_yr_df, 
        goog_earnings_10_yr_df, 
        nflx_earnings_10_yr_df, 
        amzn_earnings_10_yr_df, 
        msft_earnings_10_yr_df
    ],
    axis=1
)
```
