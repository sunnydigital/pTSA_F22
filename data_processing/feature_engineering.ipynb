```python
import numpy as np
import pandas as pd
```

```python
# importing stock price df
history = pd.read_csv('../scraping_data/SP_df_historical_daily.csv')
```

```python
# reorienting history df, selecting a singular price (closing price), cleaning up extra columns, dropping rows with more than 5% of nan values
history = history.transpose()
history.rename(columns=history.iloc[0], inplace=True)
history.columns.values[0] ='Time'
history.drop(columns='Date', axis=1, inplace=True)

history.reset_index(inplace=True)
history.columns.values[0] ='Symbol'
history['Symbol'] = history['Symbol'].apply(lambda x: x[:-2])
history = history[history['Time'] == 'Close']
history.drop(columns='Time', axis=1, inplace=True)

history = history[history.isnull().sum(axis=1) / history.shape[1] < 0.05]
history
```

```python
# inporting ticker info df
info =  pd.read_csv('../scraping_data/SP_df.csv')
info
```

```python
df = history.merge(info[['Symbol', 'GICS Sector', 'GICS Sub-Industry']], how='left', left_on='Symbol', right_on='Symbol')
cols = df.columns.to_list()[-2:] + df.columns.to_list()[:-2]
df = df[cols]
df
```

```python
# for averaging purposes, checking the number of SP500 companies in each sector
# since all sectors contain sufficient companies, we might be able to use a refined categorization of sub-indiustries (under sector)
df.groupby('GICS Sector').count()['Symbol']
```

```python
# for averaging purposes, getting the sub industries with sufficient numbers (>=10) of SP500 companies
sub_industries_series = df.groupby('GICS Sub-Industry').count()['Symbol']
sub_industries = sub_industries_series[sub_industries_series >= 10].keys().to_list()
sub_industries
```

```python
# geting the taregt company histories
df = df[df['GICS Sector'] == 'Information Technology']
# df.drop(columns='GICS Sub-Industry', axis=1, inplace=True)
df.reset_index(inplace=True)
df.drop(columns='index', axis=1, inplace=True)
df
```
